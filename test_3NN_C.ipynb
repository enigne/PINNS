```python
import sys
import os
import tensorflow as tf
import numpy as np
sys.path.append(".")
sys.path.append("./utils")
from custom_lbfgs import *
from SSAutil import *
from neuralnetwork import NeuralNetwork, MinmaxScaleLayer, UpScaleLayer
from SSA_equations.SSANN import SSA3NN_calvingfront_invertC
from logger import Logger
from plotting import *
import matplotlib.pyplot as plt
```

```python
# Manually making sure the numpy random seeds are "the same" on all devices
np.random.seed(1234)
tf.random.set_seed(1234)

# Hyper parameters {{{
hp = {}
# Data size on the solution u
hp["N_u"] = 3000
# Collocation points size, where weâ€™ll check for f = 0
hp["N_f"] = 500
# DeepNN topology (2-sized input [x t], 8 hidden layer of 20-width, 1-sized output [u]
hp["layers"] = [2, 20, 20, 20, 20, 20, 2]
hp["h_layers"] = [2, 20, 20, 20, 20, 20, 2]
hp["C_layers"] = [2, 20, 20, 20, 20, 20, 1]
# Setting up the TF SGD-based optimizer (set tf_epochs=0 to cancel it)
hp["tf_epochs"] = 100000
hp["tf_lr"] = 0.001
hp["tf_b1"] = 0.99
hp["tf_eps"] = 1e-1
# Setting up the quasi-newton LBGFS optimizer (set nt_epochs=0 to cancel it)
hp["nt_epochs"] = 50000
hp["nt_lr"] = 0.8
hp["nt_ncorr"] = 50
hp["log_frequency"] = 1000
hp["use_tfp"] = True
# Record the history
hp["save_history"] = True
hp["history_frequency"] = 10
# path for loading data and saving models
repoPath = "./"
appDataPath = os.path.join(repoPath, "matlab_SSA", "DATA")
# path = os.path.join(appDataPath, "SSA2D_nocalving.mat")
# path = os.path.join(appDataPath, "SSA2D_seg_nocalving.mat")
path = os.path.join(appDataPath, "SSA2D_segCF.mat")
# path = os.path.join(appDataPath, "Helheim_Weertman_iT080_PINN_fastflow.mat")
modelPath = "./Models/test"
reloadModel = False # reload from previous training
#}}}
```

```python
# load the data
x, y, Exact_vx, Exact_vy, X_star, u_star, X_u_train, u_train, X_f, X_bc, u_bc, X_cf, n_cf, xub, xlb, uub, ulb = prep_Helheim_data_all(path, hp["N_u"], hp["N_f"])

# Creating the model and training
logger = Logger(hp)
pinn = SSA3NN_calvingfront_invertC(hp, logger, X_f, 
        X_bc, u_bc,
        X_cf, n_cf,
        xub, xlb, uub, ulb, 
        modelPath, reloadModel,
        mu=1.1060e8,
        loss_weights=[1e-1, 1e-2, 1e-2, 1e-4, 1e-9])

X_u = pinn.tensor(X_star)
u = pinn.tensor(u_star)
# error function for logger
def error():
    return pinn.test_error(X_u, u)
logger.set_error_fn(error)
```

```python
# train the model
pinn.fit(X_u_train, u_train)

# save
#pinn.model.save("./Models/SSA2D_friction_1e_4_TF"+str(hp["tf_epochs"]) +"_NT"+str(hp["nt_epochs"]))
```

```python
import matplotlib.pyplot as plt

# Define the number of neurons in each layer
input_layer = 3
hidden_layer1 = 5
hidden_layer2 = 5
hidden_layer3 = 5
output_layer = 1

# Define the figure size
fig, ax = plt.subplots(figsize=(8, 8))

# Draw the input layer
for i in range(input_layer):
    ax.add_artist(plt.Circle((0, i+1), radius=0.2, color='blue'))

# Draw the hidden layers
for i in range(hidden_layer1):
    ax.add_artist(plt.Circle((1, i), radius=0.2, color='green'))
for i in range(hidden_layer2):
    ax.add_artist(plt.Circle((2, i), radius=0.2, color='green'))
for i in range(hidden_layer3):
    ax.add_artist(plt.Circle((3, i), radius=0.2, color='green'))

# Draw the output layer
for i in range(output_layer):
    ax.add_artist(plt.Circle((4, i+2), radius=0.2, color='red'))

# Draw the connections between the layers
for i in range(input_layer):
    for j in range(hidden_layer1):
        ax.add_artist(plt.Line2D((0, 1), (i+1, j), color='black', linewidth=0.1))
for i in range(hidden_layer1):
    for j in range(hidden_layer2):
        ax.add_artist(plt.Line2D((1, 2), (i, j), color='black', linewidth=0.1))
for i in range(hidden_layer2):
    for j in range(hidden_layer3):
        ax.add_artist(plt.Line2D((2, 3), (i, j), color='black', linewidth=0.1))
for i in range(hidden_layer3):
    for j in range(output_layer):
        ax.add_artist(plt.Line2D((3, 4), (i, j+2), color='black', linewidth=0.1))

# Set the axis limits and labels
ax.set_xlim([-1, 5])
ax.set_ylim([-1, max(input_layer, hidden_layer1, hidden_layer2, hidden_layer3, output_layer)])
ax.set_aspect('equal')
ax.set_xticks(range(5))
ax.set_xticklabels(['Input', 'H1', 'H2', 'H3', 'Output'])
ax.set_title('Deep Neural Network Schematic')

# Show the plot
plt.show()
```
